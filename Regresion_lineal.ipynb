```
#Importando librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```
data = pd.read_excel("datos_producci√≥n.xlsx")
data.head(2)
```

```
#Declaramos las variables que utilizaremos
variable_x = 'Horas Trabajadas'
variable_y = 'Productos Terminados'
```

```
#Generamos nuestro analisis
modelo = LinearRegression()
modelo.fit(data[[variable_x]], data[variable_y])
print('Ecuacion de la recta: y = ', round(modelo.coef_[0],3), 'x + ', round(modelo.intercept_,3))#Obtenemos la ecuacion de la recta
print('Coeficiente de correlacion: ', round(np.corrcoef(data[variable_x], data[variable_y])[0,1], 3))#Obtenemos el coeficiente de correlacion
print('Coeficiente de determinacion: ', round(r2_score(data[variable_y], modelo.predict(data[[variable_x]])), 3))#Obtenemos el coeficiente de determinacion

#Graficamos el intervalo de confianza
sns.regplot(x = data[variable_x], y = data[variable_y], ci = 95, line_kws = {"color" : "black"}, scatter_kws = {"color" : "blue"})
plt.title("Regresion lineal simple con intervalo de confianza del 95%")
plt.show()
```
